```python
import m750; reload(m750)
from __future__ import division
from collections import defaultdict

import pytagcloud; reload(pytagcloud)
import IPython.display

text, entries = m750.read_local_750words(r'U:\Downloads')
wfunc = lambda x: x.lower() if len(x) > 2 else ''
es = m750.entrystats(entries, wordfunc=wfunc)
```

```python
content_ratios = []
cr_sums = {}
for e in es[::-1]:
    entry_content_ratio = defaultdict(float)
    for word, freq in e.freqdict.items():
        if word == '':
            continue
        entry_content_ratio[word] = freq / sum(e.freqdict.values())
        if not word in cr_sums:
            cr_sums[word] = 0
    content_ratios.append(entry_content_ratio)
```

```python
for word in cr_sums:
    cr_sums[word] = sum((cr[word] / len(es) for cr in content_ratios))
cr_wfs = m750.wordfreqpairs(cr_sums)
stops = [word for word, prop in cr_wfs[:50]]
print stops
print [word for word, prop in cr_wfs[60:100]]
```
